```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("../input/up-crime-data-year-2019/MR data - Compiled Data Set.csv")
df.head(15)
```

```python
df = df.drop(['Title','Text'], axis=1)
df = df.fillna(0)
df.columns = [i for i in range(len(df.columns))]
df.head(15)
```

```python
d = {0:0,'Property Disputes':1,'Family Dispute':2,'Petty Quarrels':3,'Money Disputes':4, 'Personal Vendetta':5, 'Love Affairs':6, 'Casteism':7,'Unknown reasons':8}
```

```python
df[1] = df[1].apply(lambda x : d[x])
df.head(20)
```

```python
df[11]=0
df[12]=0
for i in range(0,len(df)):
    #print(df.iloc[i,8])
    if ',' in str(df.iloc[i,8]):
        cl=df.iloc[i,8].split(',')
        df.iloc[i,8]=int(cl[0])
        df.iloc[i,11]=int(cl[1])
        if len(cl)>2:
            df.iloc[i,12]=int(cl[2])
    elif int(df.iloc[i,8])>100:
        df.iloc[i,11]=int(df.iloc[i,8])%100
        df.iloc[i,8]=int(df.iloc[i,8])//100
```

```python
df.head(15)
```

```python
arr1 = np.array(df[0])
plt.hist(arr1)
plt.title("No. of victim v/s city")
plt.xlabel("City")
plt.ylabel("No. of victims")
plt.show()
```

```python
count=[0]*11;
for i in range(len(df)) :
    for j in range(1,11) :
        if int(df.iloc[i,8])==(j) or int(df.iloc[i,11])==(j) or int(df.iloc[i,12])==(j) :
            count[j]=count[j]+df.iloc[i,9]+df.iloc[i,10]
del count[0]
reason = list(("Murder with Rape","Dowry Deaths","Suicide","Kidnapping","Acid Attack","Cruelty by Husband/in-laws","Rape only","Assault on Women","Cyber Crimes","Sexual assault onchildren")) 
plt.figure(figsize=(19,5))
plt.bar(reason, count, width = 0.4)
plt.title("No. of victim v/s reason")
plt.xlabel("Reason")
plt.ylabel("No. of Victims")
```

```python
x=df.groupby(by=0).sum()
x1=np.arange(len(x))
x2=[i+0.2 for i in x1]
x3=[i+0.2 for i in x2]
plt.bar(x1,x.iloc[:,5],0.4,label="female")
plt.bar(x2,x.iloc[:,4],0.4,label="child")
plt.bar(x3,x.iloc[:,6],0.4,label="male")
plt.xlabel("City")
plt.ylabel("No. of Kidnapping")
plt.title("City Vs Kidnapping")
plt.xticks(x1+0.1,x.index)
plt.legend()
plt.show()
```
